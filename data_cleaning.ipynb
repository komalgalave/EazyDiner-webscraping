```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import re
```

```python
df = pd.read_csv('eazydiner_dataset.csv')
```

```python
df
```

```python
df['Area'] = df['Area'].str.replace('\xa0 \| View All\(\d+\) Outlets', '', regex=True)
```

```python
df
```

```python
# Remove everything after the first comma in 'Area'
df['Area'] = df['Area'].str.split(',', n=1).str[0]
```

```python
df
```

```python
df['Amount'] = df['Amount'].str.replace('\D', '', regex=True)
```

```python
df
```

```python
locations = ['jaipur', 'goa', 'patna', 'mumbai', 'bengaluru']
loc = np.repeat(locations, 135)
df['Location'] = loc
```

```python
df
```

```python
Restaurant_Discount = []
for i in df['Discount']:
    if '+' in i:
        d = i.split('+')[0]
        Restaurant_Discount.append(d)
    else:
        Restaurant_Discount.append(np.nan)
df['Restaurant_Discount'] = Restaurant_Discount

Restaurant_Discount = []
for i in df['Restaurant_Discount']:
    if isinstance(i, str) and 'Off' in i:
        d = re.sub('\D', '', i)
        Restaurant_Discount.append(d)
    else:
        Restaurant_Discount.append(np.nan)
df['Restaurant_Discount'] = Restaurant_Discount
```

```python
df
```

```python
Payment_Discount = []
for i in df['Discount']:
    if '+' in i:
        d = i.split('+')[1]
        Payment_Discount.append(d)
    else:
        Payment_Discount.append(np.nan)
df['Payment_Discount'] = Payment_Discount

Payment_Discount = []
for i in df['Payment_Discount']:
    if isinstance(i, str) and 'Off' in i:
        d = re.sub('\D', '', i)
        Payment_Discount.append(d)
    else:
        Payment_Discount.append(np.nan)
df['Payment_Discount'] = Payment_Discount
```

```python
df
```

```python
df = df.drop(columns=['Discount'])
```

```python
df
```

```python
df['Restaurant_Discount'] = df['Restaurant_Discount'].fillna(0)
df['Payment_Discount'] = df['Payment_Discount'].fillna(0)
```

```python
df
```

```python
# Convert to string first, remove non-numeric characters, then to int
df['Amount'] = df['Amount'].astype(str).str.replace(r'[^0-9]', '', regex=True).astype(int)
df['Restaurant_Discount'] = pd.to_numeric(df['Restaurant_Discount'].astype(str).str.replace(r'[^0-9]', '', regex=True), errors='coerce').fillna(0).astype(int)
df['Payment_Discount'] = pd.to_numeric(df['Payment_Discount'].astype(str).str.replace(r'[^0-9]', '', regex=True), errors='coerce').fillna(0).astype(int)
```

```python
df
```

```python
df['discount_value'] = (df['Restaurant_Discount'] + df['Payment_Discount'])/100 * df['Amount']
df['final_bill'] = df['Amount'] - df['discount_value']

df['discount_value'] = df['discount_value'].map('{:.2f}'.format)
df['final_bill'] = df['final_bill'].map('{:.2f}'.format)
```
